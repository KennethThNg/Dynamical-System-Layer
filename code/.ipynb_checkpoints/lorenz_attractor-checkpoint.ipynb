##Â Set up of Lorenz attractor
This is the next step of the dynamical system layer benchmark, the first part of the benchmark is contained in the notebook ``pendulum.ipynb`` where the linear dynamical system has been validated. For this step, we add a non-quadratic part in differential equations and we obtain
$$\dot{y}(t)=Ay(t) + \phi(y(t)),y(0)=y_0,$$
where $\phi(y)$ denotes the non-linear part. The Lorenz attractor is a $3-$dimension dynamical system $u(t)=(x(t), y(t), z(t))$ described by
\begin{align*}
&\dot{x}(t)= \sigma(y(t)-x(t)),\\
&\dot{y}(t)= x(t)(\rho-z(t))-y(t),\\
&\dot{z}(t)= x(t)y(t) - \beta z(t),
\end{align*}
with initial condition $u(0)=u_0=(x_0, y_0, z_0).$ We can rewrite the Lorenz attractor dynamical system as
$$\dot{u}(t)=Au(t)+\phi(u(t)),u(0)=u_0,$$
where $A$ is a $(3\times 3)-$matrix defined by
$$A=
\begin{pmatrix}
-\sigma &\sigma &0\\
\rho &-1 &0\\
0 &0 &-\beta
\end{pmatrix}
$$
and $\phi(u)=(\phi_1(u), \phi_2(u), \phi_3(u))^{*}$ where $\phi_i(u)=u^*B_iu$ and $B_i$ a $(3\times 3)-$matrix, then $B$ is a $(3\times 3\times 3)-$matrix defined by
$$
B=
\begin{pmatrix}
\begin{pmatrix}
0 &0 &0\\
0 &0 &0\\
0 &0 &0\\
\end{pmatrix}\\
\begin{pmatrix}
0 &0 &-1\\
0 &0 &0\\
0 &0 &0\\
\end{pmatrix}\\
\begin{pmatrix}
0 &1 &0\\
0 &0 &0\\
0 &0 &0
\end{pmatrix}
\end{pmatrix}
$$
With the Euler forward method, we obtain the approximation with $u_n=u(t_n)$ defined by
\begin{align*}
u_{n+1} &= u_n + hAu_n + h\phi(u_n),\\
&= (I+hA)u_n + h\phi(u_n),\\
&= H_1u_n + \phi_h(u_n)
\end{align*}
where the matrix $H_1$ is defined by 
$$H_1=
\begin{pmatrix}
1-h\sigma &h\sigma &0\\
h\rho &1-h &0\\
0 &0 &1-h\beta
\end{pmatrix}
$$
and $\phi_h(u)=u^{*}H_2u$ with $H_2=hB.$

```python
# Simulation of the Lorenz attractor
in_dim = 3
x = 0
y = 1
z= 1.05
sigma = 10
rho = 30
beta = 8/3
dt = 0.01
T = 10000

sol = [[[x, y, z]]]

H1 = torch.Tensor([[1-dt*sigma, dt*sigma, 0], [dt*rho, 1-dt, 0], [0, 0, 1-dt*beta]])
Q0 = np.asarray([[0,0,0], [0,0,0], [0,0,0]])*dt
Q1= np.asarray([[0,0,-1], [0,0,0], [0,0,0]])*dt
Q2 = np.asarray([[0,1,0],[0,0,0], [0,0,0]])*dt
B = np.array([Q0, Q1, Q2])
H2 = torch.Tensor(B)

W = {'linear': H1, 'quadratic': H2}

ode_true = NNODEModel(in_dim, in_dim, in_dim, weight=W, reset=False)

for t in range(T):
    u = torch.Tensor(sol[-1])
    pred = ode_true(u)
    sol.append(pred.detach().tolist())
    
X = np.array(sol)[:-1].squeeze(1)
Y = np.array(sol)[1:].squeeze(1)

fig = plt.figure(figsize=(20, 10))
ax = fig.gca(projection='3d')
ax.plot(X[:,0], X[:,1], X[:,2])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Lorenz attractor')
plt.show()
```

## Training model
We consider the general dynamical systems defined by
$$u_{n+1}=W_1u_n + u_n^{*}W_2u_n,$$
where $W_1$ is a $(3\times 3)-$weight and $W_2$ is a $(3\times 3\times 3)-$weight. We determine if $W_1$ converges to $H_1$ and $W_2$ converges to $H_2.$

```python
# Convert data to torch.Tensor
train_x = torch.Tensor(X)
train_y = torch.Tensor(Y)
```

```python
# Init model
ode_train = NNODEModel(in_dim, in_dim, in_dim)
model = NeuralNetRegressor(module=ode_train, criterion=nn.MSELoss, optimizer=optim.Adagrad, lr=1e-2, batch_size=100, max_epochs=5000)
```

```python
# Train model
model.fit(train_x, train_y)
```

```python
#plot the prediction
prediction = model.predict(train_x)
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.plot(X[:,0], X[:,1], X[:,2], label='Ground truth');
ax.plot(prediction[:,0], prediction[:,1], prediction[:,2], label='Prediction')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Prediction of the harmonic oscillator')
ax.legend()
plt.show()
```

```python
# Comparison between 
print('Real linear weight:')
print(W['linear'])
print('Real quadratic weight:')
print(W['quadratic'])
print('----------------------------------')
print('linear weight estimation:')
pred_batch_param = ode_train.state_dict()
print(pred_batch_param['lin_ode.weight'])
print('quadratic weight estimation:')
print(pred_batch_param['nl_ode.weight'])
```

We observe that the model converges to the dynamical system describing the Lorrenz attractor. We directly train the model on batches of size $100$ to reduce the cost of computation,. Each step of the benchmark is the same as with the harmonic oscillator, but we skip the training on the all dataset due to large computational cost. Then we can validate the quadratic model.
