# Case Study I: Harmonic oscillator pendulum

This notebook contains the full pipeline of the dynamical system layer implementation for the harmonic oscillator. The aim is to benchmark this layer by starting with this simplest case.

## Import libraries
First, we import the ``DynamicalSystem`` module.

```python
%load_ext autoreload
%autoreload 2
```

```python
# import the 'DynamicalSystem' module
import DynamicalSystem
from DynamicalSystem.ODEModel import LinearODEModel
from skorch import NeuralNetRegressor
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

## Set up the harmonic oscillator simulation
This is the first step of the benchmark for the dynamical system layer. We start first with the harmonic oscillator defined by the 2-dimension differential equations
$$\dot{y}(t) = Ay(t),y(0)=(x_0, v_0)$$
where $y=(x, \dot{x})$ with $x$ describing the harmonic motion of the oscillator defined by $\ddot{x}=-kx$ with $x(0)=x_0$ and $\dot{x}(0)=v_0.$ The matrix $A$ is defined by
$$A=
\begin{pmatrix}
 0 & 1\\
-k & 0
\end{pmatrix}
$$
We use the forward Euler method to simulate the data
$$y_{n+1}=y_n + hAy_n=(I+hA)y_n=Hy_n,$$
where $H$ defined by with $h=1,$
$$H=
\begin{pmatrix}
 1 & 1\\
-k & 1
\end{pmatrix}
$$


```python
# Simulation of the harmonic oscillator
in_dim = 2
sol = [[[0.99,1]]]
T = 100
k = 0.01
W = torch.Tensor([[1,1], [-k, 1]])
ode_true = LinearODEModel(in_dim, in_dim, weight=W, reset=False)

for t in range(T):
    x = torch.Tensor(sol[-1])
    pred = ode_true(x)
    sol.append(pred.detach().tolist())
    
X = np.array(sol)[:-1].squeeze(1)
Y = np.array(sol)[1:].squeeze(1)

plt.plot(X[:,0], label='Input');
plt.plot(Y[:,0], label='Target')
plt.title('Harmonic oscillator')
plt.xlabel('Time')
plt.ylabel('Pendulum data')
plt.legend()
plt.show()
```

##Â Training model
Let's consider the first linear dynamical system defined by
$$y_{n+1}=Wy_n,$$
where $W$ is the weight matrix to learn.

We determine if the weight $W$ converges to $H.$

```python
#Convert data to torch.Tensor
train_x = torch.FloatTensor(X)
train_y = torch.FloatTensor(Y)
```

```python
# Init model
ode_train = LinearODEModel(in_dim, in_dim)
model = NeuralNetRegressor(module=ode_train, criterion=nn.MSELoss, optimizer=optim.SGD, lr=0.01, batch_size=train_x.shape[0], 
                           max_epochs=600)
```

```python
#Train model
model.fit(train_x, train_y)
```

```python
#plot the prediction
prediction = model.predict(train_x)
plt.plot(Y[:,0], label='Ground truth');
plt.plot(prediction[:,0], label='Prediction')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Pendulum')
plt.title('Prediction of the harmonic oscillator')
plt.show()
```

```python
# Comparison between 
print('Real weight:')
print(W)
print('----------------------------------')
print('weight estimation:')
pred_param = ode_train.state_dict()
print(pred_param['ode.weight'])
```

We observe that model converges to the harmonic oscillator, then we validate the first step. The next step is too complexify the model, first we add a bias to the dynamical system layer, then we obtain the model
$$y_{n+1}= b + Wy,$$
where $b$ is the bias vector and $W$ is the weight matrix. If we train the model, we expect that $b$ converges to the zeros vector and $W$ converges to the matrix $H.$

```python
# Implement the bias model 
ode_train_bias = LinearODEModel(in_dim, in_dim, bias=True)
model_bias = NeuralNetRegressor(module=ode_train_bias, criterion=nn.MSELoss, optimizer=optim.SGD, lr=0.01, batch_size=train_x.shape[0],
                               max_epochs=600)

model_bias.fit(train_x, train_y)
```

```python
#plot the prediction
prediction_bias = model_bias.predict(train_x)
plt.plot(Y[:,0], label='Ground truth');
plt.plot(prediction_bias[:,0], label='Prediction')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Pendulum')
plt.title('Prediction with bias of the harmonic oscillator')
plt.show()
```

```python
# Comparison between 
print('Real weight:')
print(W)
print('----------------------------------')
print('weight estimation:')
pred_bias_param = ode_train_bias.state_dict()
print(pred_bias_param['ode.weight'])
print('bias estimation:')
print(pred_bias_param['ode.bias'])
```

We observe that the weight converges to the pendulum matrix and the bias is approximatively $0.$

In the previous step, we trained the model on the entire dataset. For the third step, we consider smaller batch to improve the training. 

```python
# init the model and train per batch of size 10
ode_train_batch = LinearODEModel(in_dim, in_dim)
model_batch = NeuralNetRegressor(module=ode_train_batch, criterion=nn.MSELoss, optimizer=optim.SGD, lr=0.01, batch_size=10, 
                                 max_epochs=600)
model_batch.fit(train_x, train_y)
```

```python
#plot the prediction
prediction_batch = model_batch.predict(train_x)
plt.plot(Y[:,0], label='Ground truth');
plt.plot(prediction_batch[:,0], label='Prediction')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Pendulum')
plt.title('Prediction with batch size 10 of the harmonic oscillator')
plt.show()
```

```python
# Comparison between 
print('Real weight:')
print(W)
print('----------------------------------')
print('weight estimation:')
pred_batch_param = ode_train_batch.state_dict()
print(pred_batch_param['ode.weight'])
```

If we split the dataset per mini-batch, the learning of the method is even better
